```python
# test_env.py
import sys
import os
from importlib.metadata import version as pkg_version
from dotenv import load_dotenv

print("环境验证开始")
print("=" * 40)

# 0. 获取虚拟环境信息（兼容 venv + Conda）
print("【虚拟环境信息】")
# 判断Conda环境
conda_env = os.getenv("CONDA_DEFAULT_ENV")
conda_prefix = os.getenv("CONDA_PREFIX")
# 判断原生venv
is_venv = hasattr(sys, 'real_prefix') or (sys.base_prefix != sys.prefix)

if conda_env and conda_prefix:
    print("✅ 当前处于 Conda 虚拟环境")
    print(f"Conda 环境名称: {conda_env}")
    print(f"Conda 环境完整路径: {conda_prefix}")
elif is_venv:
    venv_path = sys.prefix
    venv_name = os.path.basename(venv_path)
    print("✅ 当前处于 venv 虚拟环境")
    print(f"虚拟环境名称: {venv_name}")
    print(f"虚拟环境完整路径: {venv_path}")
else:
    print("❌ 当前未激活任何虚拟环境，使用系统原生Python")
print("-" * 40)

# 1. Python 版本
print(f"Python 版本: {sys.version}")
print(f"Python 安装根目录: {sys.base_prefix}")
print("-" * 40)

# 2. 关键包版本
print("【依赖包版本】")
print(f"LangGraph 版本: {pkg_version('langgraph')}")

import langchain_core
print(f"LangChain Core 版本: {langchain_core.__version__}")

import pydantic
print(f"Pydantic 版本: {pydantic.__version__}")
print("-" * 40)

# 3. .env 加载
print("【密钥配置检测】")
load_dotenv(override=True)
deepseek_key = os.getenv("DEEPSEEK_API_KEY")
if deepseek_key:
    print(f"DeepSeek API Key 已配置: {deepseek_key[:8]}...{deepseek_key[-4:]}")
else:
    print("⚠️  未检测到 DEEPSEEK_API_KEY，请在 .env 文件中配置")
print("-" * 40)

# 4. 模型调用测试
print("【模型连通性测试】")
if deepseek_key:
    from langchain_deepseek import ChatDeepSeek
    model = ChatDeepSeek(model="deepseek-v4-flash")
    response = model.invoke("回复'一切正常'四个字，不要其他内容")
    print(f"模型调用测试: {response.content}")
else:
    print("⚠️  跳过模型调用测试（缺少 API Key）")

print("=" * 40)
print("环境验证完成")
print("=" * 40)
```
