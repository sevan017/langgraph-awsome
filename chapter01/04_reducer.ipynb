## 1.4 reducer 状态合并定义

代码很简单，并不涉及langgraph的内容，但langgraph底层就是这样合并状态的

```python
# left 当前状态（从最开始位置合并到当前位置的值）
# right 新状态（当前节点的值）
# 返回合并后的状态（值）
def my_reducer(left: list[str], right: list[str]) -> list[str]:
    # 合并状态，python中 + 运算符求并集
    return left + right

# 1、示例， 合并状态
left  = ["start", "node_1 运行完毕"]
right = ["node_2 运行完毕"]

# 合并状态
merged = my_reducer(left, right)
print(merged)


```
